```python
import random as r
x=["P1","P2","P3","P4","P5","P6","P7","P8","P9","P10","P11","P12","P13","P14","P15"]
y=[]
for i in range(15):
 y.append(r.randint(10,300))
for i in range(2,32768):
 product_f=[]
 n=r.randint(2,15)
 f=r.sample(y,n)
 if sum(f)==500:
   print(f)
   for i in f:
     m=y.index(i)
     product_f.append(x[m])
     print(product_f)
```
